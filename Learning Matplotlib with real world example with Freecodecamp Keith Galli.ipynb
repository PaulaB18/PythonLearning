# FIFA DATA

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

## Using histogram to see the skill levels of the players

```python
bins = [40,50,60,70,80,90,100] #This to set up the parameter of x axis

plt.hist(fifa.Overall, bins=bins, color='#dbccd3')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('The Distribution of FIFA Players Skill Level')

plt.show()
```

## Using Pie Chart to See The Comparison of Preferred Foot Used by FIFA Players

```python
#First, using loc to calculate the total number of players who used right or left foot.
#loc is for filtering the data with the specific condition, like below the variable left will only calculate if the preferred foot is left
#We must use [0] after count(), because if we only use count(), it will give us an array instead of total number.

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0] 
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left','Right'] #We will pass this variable for the labels in the pie chart

colors = ['#f5e1ed','#aed1d1']

#Now make a pie chart by passing these two variables left and right above:

plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f %%') #autopct is to give an automated percentage to each section.

#Now give a title
plt.title('Preferred Foot Used by FIFA Players Comparison')

plt.show()
```

## Using Pie Chart to See the Weight Ranges of FIFA Players

```python
#In weight column, the data type is still string and has 'lbs' attached on it, so the first step is to change to integer

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

#The strip() method removes both the leading and the trailing spaces.In the case above, the x.strip('lbs') is to remove lbs
#The int method in int(x.strip('lbs')) is to convert the type from string to int

plt.style.use('ggplot')


#Now make the weight categories:
light = fifa.loc[fifa.Weight<125].count()[0]
light_medium = fifa.loc[(fifa.Weight>=125) & (fifa.Weight<150)].count()[0]
medium = fifa.loc[(fifa.Weight>=150) & (fifa.Weight<175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight>=175) & (fifa.Weight<200)].count()[0]
heavy = fifa.loc[fifa.Weight>=200].count()[0]

#Create variable that we are going to pass to the plt.pie
weights = [light, light_medium, medium, medium_heavy, heavy]

#Create labels
labels = ['Under 125','125-150','150-175','175-200','Above 200']

#Using explode to give a space between each sections in the pie
explode = [0.5,0,0,0,0.5]

#Now create a pie chart using weights variable

plt.pie(weights, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
# pctdistance is to give the distance of percentage from the center. 


plt.show()


```

## Whisker Chart

```python
#First, we need to create variable that we will pass to the chart
Barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
Juventus = fifa.loc[fifa.Club == 'Juventus']['Overall']
```
